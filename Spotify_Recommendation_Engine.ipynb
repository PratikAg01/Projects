```python
from sklearn.preprocessing import MinMaxScaler
datatypes=['int16','int32','int64','float16','float32','float64']
normalization=spotify_df.select_dtypes(include=datatypes)
for col in normalization.columns:
    MinMaxScaler(col)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=10)
features = kmeans.fit_predict(normalization)
```

```python
spotify_df['features']=features
```

```python
from tqdm import tqdm 
```

```python
class Spotify_recommendation():
	def __init__(self,dataset):
		self.dataset=dataset
	def recomend(self,songs,amount=1):
		distance=[]
		song=self.dataset[(self.dataset.track_name.str.lower()==songs.lower())].head(1).values[0]
		rec=self.dataset[self.dataset.track_name.str.lower()!=songs.lower()]
		for songs in tqdm(rec.values):
			d=0
			for col in np.arange(len(rec.columns)):
				if not col in[1,2,3,4,20]:
					d=d+np.absolute(float(song[col])-float(songs[col]))
			distance.append(d)
		rec['distance']=distance
		rec=rec.sort_values('distance')
		columns=['artists','track_name']
		return rec[columns][:amount]
recommendation=Spotify_recommendation(spotify_df)
recommendation.recomend("Lovers Rock",10)
```
